```python
import cv2
import os
```

```python
PATH_TO_DATASET = 'datasets/a9_dataset_r01_s02/_images/s040_camera_basler_north_16mm'
file_list = os.listdir(PATH_TO_DATASET)

sample_img = os.path.join(PATH_TO_DATASET, file_list[0])
sample_img = cv2.imread(sample_img)

height, width = sample_img.shape[:2]

fourcc = cv2.VideoWriter.fourcc(*'avc1')
out = cv2.VideoWriter('tumtraf_sample_video.mp4', fourcc, 30, (width, height))
count = 0


for filename in file_list:
  frame = cv2.imread(os.path.join(PATH_TO_DATASET, filename))
  out.write(frame)
  count += 1
  
  if count == 150:
    break
  
out.release()
```
